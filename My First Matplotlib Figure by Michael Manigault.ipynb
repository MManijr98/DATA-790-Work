```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

x = np.array(range(1, 100))
y = np.array(x**2)

fig, ax = plt.subplots(1, 2)

ax[0].plot(x,x, color = 'green', linestyle = '--')
ax[1].plot(x, 5**x, 'b-')

ax[0].set_xlabel('x')
ax[0].set_ylabel('x')

ax[1].set_xlabel('x')
ax[1].set_ylabel('$5^x$')
fig.suptitle('A simple plot', fontsize=14)
```
