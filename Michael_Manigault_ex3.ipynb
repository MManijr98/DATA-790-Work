```python
import seaborn as sns  
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_finalproject = pd.read_excel('MLB Diversity V2.xlsx', sheet_name=2)
```

```python
df_finalproject
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.kdeplot(df_finalproject.of_players, ax=ax,                 
                     bw=0,               
                     shade=True,          
                     alpha=0.6
           )

ax.set_xlabel('players')
ax.set_ylabel('density')
ax.set_title('Amount of players in the MLB from 1902 to 1947')

ax.grid(axis='y')
ax.set_axisbelow(True)
plt.box(False)  
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.kdeplot(df_finalproject.of_players, ax=ax,                 
                     bw=1,               
                     shade=True,          
                     alpha=0.6
           )

ax.set_xlabel('players')
ax.set_ylabel('density')
ax.set_title('Amount of players in the MLB from 1902 to 1947')

ax.grid(axis='y')
ax.set_axisbelow(True)
plt.box(False)  
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

sns.kdeplot(df_finalproject.of_players, ax=ax,                 
                     bw=2,               
                     shade=True,          
                     alpha=0.6
           )

ax.set_xlabel('players')
ax.set_ylabel('density')
ax.set_title('Amount of players in the MLB from 1902 to 1947')

ax.grid(axis='y')
ax.set_axisbelow(True)
plt.box(False)  
```

Out of all of the bandwidths used (0, 1 and 2), the bandwidth of 0 provides a more detailed graph depicting the amount of players in the MLB from 1902 to 1947. The amount of players in the dataset is measured in percentages in decimal form. As a result, the bandwidth should be zero due to it being greater than -1 and less than 1.
