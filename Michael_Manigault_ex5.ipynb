```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline 
```

```python
url = 'https://raw.githubusercontent.com/RamiKrispin/coronavirus/master/csv/coronavirus.csv'

df = pd.read_csv(url, header = 0,
                usecols = ['date', 'province', 'country', 'type', 'cases']) 
```

```python
df_world = df[(df.type=='confirmed') & 
                df.country.isin(['United Kingdom', 'Argentina', 'Italy', 'Colombia', 'Mexico', 
                                'Peru', 'Germany', 'Iran', 'Poland',
                                'South Africa'])
              & (df['date']>='2020-01-01') & (df['date']<='2020-11-01')]
             
```

```python
s = sns.catplot(x="country", y="cases", 
                color = 'gray',
                width =0.4,
                kind='box',
                data=df_world)

s.fig.set_figwidth(10)
s.fig.set_figheight(8)
s.ax.set_title("Confirmed COVID Cases in Different Countries from January to November")
```
